```python
import pandas as pd

# Load the first dataset
remit_corridor = pd.read_csv(r'C:\Users\clint\Desktop\RER\data\Remittance_4\Monthly remittance - Corridor data.csv')

# Load the second dataset
df12 = pd.read_csv(r'C:\Users\clint\Desktop\RER\Code\12.csv')

remit_corridor.head(), df12.head()
```

```python
# Remove columns 'wb1', 'wb2', 'iso1', and 'iso2' from remit_corridor
drop_cols = ['wb1', 'wb2', 'iso1', 'iso2']
remit_corridor = remit_corridor.drop(columns=drop_cols, errors='ignore')

# Display the updated yearly remit_corridor
print("Updated remit_corridor (yearly aggregated):")
print(remit_corridor.head(10))
print(f"\nColumns: {remit_corridor.columns.tolist()}")
print(f"Shape: {remit_corridor.shape}")
remit_corridor
```

```python
# Convert monthly data to yearly by aggregating
# Group by all columns except 'month', 'received', 'sent', and 'period'
# Then sum the 'received' and 'sent' columns

# First, let's check the column names to understand the structure
print("Columns in remit_corridor:")
print(remit_corridor.columns.tolist())
print("\nSample data:")
print(remit_corridor.head())
```

```python
df12
```

```python
# Convert monthly data to yearly by aggregating
# Group by country1, country2, and year, then sum received and sent columns
remit_corridor_yearly = remit_corridor.groupby(['country1', 'country2', 'year']).agg({
    'received': 'sum',
    'sent': 'sum'
}).reset_index()

# Drop the month and period columns are automatically excluded from groupby
print("Yearly aggregated data:")
print(remit_corridor_yearly.head(10))
print(f"\nShape of original data: {remit_corridor.shape}")
print(f"Shape of yearly data: {remit_corridor_yearly.shape}")

# Update the remit_corridor to be the yearly version
remit_corridor = remit_corridor_yearly
```

```python
# Display the final formatted remit_corridor
print("Final remit_corridor dataframe (formatted like df12):")
print(f"Shape: {remit_corridor.shape}")
print("\nSample of the data:")
print(remit_corridor.head(15))

# Show some examples to verify the transformation logic
print("\n" + "="*60)
print("Verification examples:")
print("- Russian Federation → Armenia (2018): $1,049.56 million")
print("- United States → Armenia (2018): $223.98 million") 
print("- Angola → Brazil (2018): $8.50 million")
print("\nThis matches the logic:")
print("- 'received' column became: Sending Country → Receiving Country")
print("- 'sent' column became: Sending Country → Receiving Country")
remit_corridor
```

```python
# Transform remit_corridor to match df12 format
# We need to create two separate dataframes: one for received and one for sent

# Create received transactions (country1 receives money, so it's the receiving country)
received_df = remit_corridor[remit_corridor['received'].notna()].copy()
received_df = received_df[['country2', 'country1', 'year', 'received']].copy()  # country2 sends to country1
received_df.columns = ['Sending Country', 'Receiving Country', 'Year', 'Value']

# Create sent transactions (country1 sends money, so it's the sending country) 
sent_df = remit_corridor[remit_corridor['sent'].notna()].copy()
sent_df = sent_df[['country1', 'country2', 'year', 'sent']].copy()  # country1 sends to country2
sent_df.columns = ['Sending Country', 'Receiving Country', 'Year', 'Value']

# Combine both dataframes
remit_corridor_formatted = pd.concat([received_df, sent_df], ignore_index=True)

# Add the standard columns to match df12 format
remit_corridor_formatted['Unit'] = 'USD millions'
remit_corridor_formatted['Source'] = 'Roland Kpodar (IMF)'
remit_corridor_formatted['Region'] = ''  # We can fill this later if needed

# Reorder columns to match df12
remit_corridor_formatted = remit_corridor_formatted[['Sending Country', 'Receiving Country', 'Year', 'Value', 'Unit', 'Source', 'Region']]

print("Transformed remit_corridor to match df12 format:")
print(f"Shape: {remit_corridor_formatted.shape}")
print("\nFirst 10 rows:")
print(remit_corridor_formatted.head(10))
print(f"\nColumns: {remit_corridor_formatted.columns.tolist()}")

# Update the remit_corridor variable
remit_corridor = remit_corridor_formatted
```
