```python
import pandas as pd

# Load stream_table_1.csv
df1 = pd.read_csv(r'C:\Users\clint\Desktop\RER\_output\data\Remittance_3\stream_table_1.csv', encoding='utf-8', engine='python')

# Load stream_table_2_cleaned.csv
df2 = pd.read_csv(r'C:\Users\clint\Desktop\RER\_output\data\Remittance_3\stream_table_2_cleaned.csv', encoding='utf-8', engine='python')

# Display the first few rows of each dataframe
print(df1.head())
print(df2.head())
```

```python
# Split the second column (column named '1') which contains country + date
# The format appears to be "Country\nYear"
df2_split = df2['1'].str.split('\n', expand=True)

# Insert the split columns back into the dataframe
df2.insert(1, 'Country', df2_split[0])  # Insert country after first column
df2.insert(2, 'Year', df2_split[1])     # Insert year after country column

# Drop the original combined column (now named '1')
df2.drop(columns=['1'], inplace=True)

# Clean up column names
df2.columns = ['Sending Country', 'Receiving Country', 'Year', 'Value', 'Unit', 'Source']

print("Updated dataframe structure:")
print(f"Shape: {df2.shape}")
print(f"Columns: {df2.columns.tolist()}")
print("\nFirst few rows:")
print(df2.head())
```

```python
# Let's also clean up the tab characters (\t) in the data for better readability
df2 = df2.replace('\t', ' ', regex=True)

print("Data after cleaning tab characters:")
print(df2.head())

print("\nUnique receiving countries:")
print(df2['Receiving Country'].unique())

print("\nUnique years:")
print(sorted(df2['Year'].unique()))
```

```python
# Forward fill missing values - replace NaN with the value from the row above
print("Before forward fill:")
print(df2.head(10))
print("\nMissing values per column before forward fill:")
print(df2.isnull().sum())

# Apply forward fill to replace missing values using the newer ffill() method
df2_filled = df2.ffill()

print("\n" + "="*50)
print("After forward fill:")
print(df2_filled.head(10))
print("\nMissing values per column after forward fill:")
print(df2_filled.isnull().sum())

# Update df2 with the filled version
df2 = df2_filled
df2
```

```python
# Save the processed dataframe to CSV
df2.to_csv('10_9.csv', index=False)
print("DataFrame saved as '10_9.csv'")
print(f"File saved with {len(df2)} rows and {len(df2.columns)} columns")
```
